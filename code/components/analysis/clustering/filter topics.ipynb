### Filter the views from PLDA (Topic Modeling)
Keep the first k ngrams for each topic
The ngrams are presented in sorted (decending order)

```python
hypo_name='cancer'
input_path='/scratch2/esadrfa/plda/view/' + hypo_name + '/'
input_path
```

```python
# %%bash 
# ls /scratch2/esadrfa/plda/view/cancer
```

```python
#https://stackoverflow.com/a/3207973/2674061
from os import walk          

viewFiles = []
for (dirpath, dirnames, filenames) in walk(input_path):
#     print(filenames)
    for f in filenames:
#         print(f)
        if(f.find('filter')>0):
            viewFiles.append(f)
    break
```

```python
clusters={}
for f in viewFiles:
    topics = []
    flg_1st = True
    with open(input_path + f, 'r') as t:   # for each file 
#         print(f)
        ngrams=[]
        for line in t:
#             print('line', line)
            if(line.find('TOPIC')==0):     # topic at the beginning of the line
                if(flg_1st):
                    flg_1st = False
                else:                  
                    topics.append(ngrams)
                    ngrams=[]
            else:
                if (not (line.find('\n') == 0 or line.find('--') == 0)):
                    ngrams.append(line.split(' ')[0])
                    
#     print(f)
#     print(type(f.split('PMID')[1].split('---')[0]),f.split('PMID')[1].split('---')[0])
    clusters[(f.split('.')[0].replace('_',''))] = topics
                
```

```python
clusters
```

```python
#https://stackoverflow.com/a/36965528/2674061
import json

print('dumping the topics in ',input_path + 'topics.txt')
with open(input_path + 'topics.txt', 'w') as file:
     file.write(json.dumps(clusters))
```
